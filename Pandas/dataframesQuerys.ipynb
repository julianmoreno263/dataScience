ITERANDO CON DATAFRAMES

```python
df.head()
```

```python
#si utilizamos un bucle for in normal para iterar este dataframe solo nos traera las etiquetas de las columnas:
for item in df:
    print(item)
```

```python
#para iterar en un dataframe utilizamos el metodo iterrows(), este metodo solo se usa teniendo un dataframe,si el resultado lo ponemos dentro de una lista podemos ver que nos retorna una lista de tuplas porque van entre parentesis todos los resultados,cada tupla tendra dos partes,la etiqueta de la fila a la que representa y los datos de la fila:
list(df.iterrows())
```

```python
#ahora, con esta informacion ya puedo hacer operaciones,por ejemplo calcular el kilometraje promedio de los carros de la lista,para esto debo primero desempaquetar la lista,osea sacar los indices de cada fila y las filas como tal para poder recorrerlas,esto lo hago utilizando un bucle for pero creando 2 variables que seran las que iteren etiquetas y filas, y dentro del bucle para calcular el kilometraje promedio debemos restar el año de creacion del vehiculo con el año actual, pero por ejemplo si el año de creacion y el actual es el mismo,la resta dara 0,por lo que al dividir por 0 para sacar el promedio da un error porque no se puede dividir por cero,entonces tambien debemos manejar este posible error(para este ejercicio el año actual del video del profe es 2020 asi que lo hacemos igual para ver los mismos resultados):
for index,row in df.iterrows():
    if(2020 - row.Año!=0):
        #aqui hago una seleccion por el index y creando una columna nueva y esto sera el promedio de km de cada auto
        df.loc[index, "Km_media"]=row.Kilometraje/(2020-row.Año)
    else:
        df.loc[index, "Km_media"]=0#aqui trato el posible error de dividir por 0

df
```

NOTA:  ESTA ES LA DOCUMENTACION DE PANDAS:  https://pandas.pydata.org/pandas-docs/stable/

vamos a ver tratamiento de datos con Pandas, el tratamiento de datos es mejorar la calidad d elos datos antes de aplicarles alguna inteligencia o logica,osea antes de operar con los datos primero debemos limpiarlos y mejorarlos, si los datos no se tratan pueden haber resultados erroneos.Por ejmplo en nuestro df hay filas con valores nulos(NaN), debemos ver que hacer con este tipo de datos,para eso primero debemos ver la informacion de nuestro set de datos con el metodo info()

```python
#con info() podemos ver las filas que tienen valores nulos y ademas el tipo de dato de cada fila.Aqui podemos ver que en la columna Kilometraje solo hay 197 filas con valores no nulos,por lo que el resto(61 para llegar a las 258 filas) tiene valores nulos,debemos decidir que hacer con esas filas nulas.
df.info()
```

```python
#pandas tiene un metodo llamada isna() que revisa dentro de una columna cuales filas seran nulas,las que tengan True sera porque son valores nulos dentro de mi columna,por lo que debemos usar este metodo pero dentro de la columna asi:
df.Kilometraje.isna()
```

```python
#este mismo metodo podemos usarlo para trernos esas 61 filas que tiene la columna Kilometraje en nulo asi:
df[df.Kilometraje.isna()]
```

```python
#ahora, si vemos estos vehiculos aparecen con kilometraje NaN porque son nuevos,entonces en vez de eliminar estas filas puedo sustituir ese NaN por el valor 0 pues son nuevos,esto lo hago utilizando el metodo fillna(), dentro del metodo debemos poner el valor de 0 que queremos darle y utilizar el metodo inplace(), este metodo es el que en si pone ese valor dentro del dataframe,porque si solo ponemos el 0 se visualiza en el resultado actual que este en pantalla pero no se reemlaza realmente en el dataframe:
df.fillna(0, inplace=True)
df
```

```python
#para corroborar que realmente se cambio los valores de NaN  a cero,pues intento ver nuevamente las filas que estaban como NaN y veremos que ya no hay filas nulas en el campo Kilometraje.
df[df.Kilometraje.isna()]
```

```python
#puedo hacer lo mismo buscando las filas que sean 0 km utilizando el metodo query(),asi:
df.query("Cero_km==True")
```

```python
#ahora, vamos a volver al dataframe original que tiene valores nulos para ver como se eliminan,por ejemplo en un caso de que no pueda sustituir ese valor nulo sino que realmente deba eliminarlos,para eso cargamos de nuevo nuestro dataframe original.
df = pd.read_csv(r'C:\Users\USER\Desktop\Python_Data_Science\Pandas\data\db.csv',sep=';',index_col=0)
df
```

```python
#para eliminar estos valores nulos de la columna Kilometraje debo usar el metodo dropna() y dentro especifico la columna  donde estan esos valores nulos con subset,no olvidar el parametro inplace=True para que ese cambio realmente se aplique en el dataframe, al ejecutar esto y borrar esas filas ya no debo tener las 258 sino menos asi:
df.dropna(subset=["Kilometraje"], inplace=True)
df
```
