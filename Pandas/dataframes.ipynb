```python
#si ahora quiero consultar pero por filas,por ejemplo solo quiero traer las 3 primeras filas,especifico el intervalo desde la primera fila que comienza en 0 y la ultima fila se especifica trayendo la fila + 1,osea,si quiero traer de la fila 0 a la 2,debo sumar 1 ,osea la fila 3 que no se traera,traera hasta la 2 pero se especifica la 3 para que python sepa que debe traer el intervalo 0:2.
df[0:3]
```

```python
#ahora, vamos a ver los metodos loc y iloc que seran los mas usados en data science porque son mas practicos,el metodo loc ,me permite buscar pero usando las etiquetas de las filas y de las columnas,por ejemplo voy a buscar los datos del auto con etiqueta de fila Passat:
df.loc[["Passat"]]#recordar que con doble [[]] formateo la salida a dataframe
```

```python
#si quiero hacer una busqueda con mas de una etiqueta,siempre debe ser con [[]],osea en formato dataframe,sino sale error:
df.loc[["Passat","Crossfox"]]
```

```python
#ahora, si solo quiero traer en los resultados la informacion de una columnas,las especifico tambien entre [],recordar que para hacer esto siempre se ponen las etiquetas de las filas primero y separadas con coma se ponen las etiquetas de las columnas asi:
df.loc[["Passat","Crossfox"],["Motor","Valor"]]
```

```python
#si quiero ver todas las filas de mis datos con solo dos columnas las puedo especificar con dos puntos asi:
df.loc[:,["Motor","Valor"]]
```

```python
#con el metodo iloc hago las mismas consultas pero utilizando los indices,no las etiquetas:
df.iloc[[1]]
```

```python
#si quiero traer un intervalo de datos lo hago entre [] sencillas asi:
df.iloc[1:4]
```

```python
#si quiero traer solo ciertas columnas y ademas mostrarlas en un orden especifico lo hago asi:
df.iloc[1:4,[0,5,2]]
```

```python
#puedo hacer lo mismo para traer las filas que yo quiera asi:
df.iloc[[1,42,22],[0,5,2]]
```

```python
#al igual que con loc,si quiero mostrar todas las filas lo hago con dos puntos asi:
df.iloc[:,[0,5,2]]
```
