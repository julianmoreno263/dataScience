Vamos a ver como manejar los datos que llegan nulos,osea no tienen  ningun valor.Aqui nosotros debemos decidir que hacer con este tipo de datos,si eliminarlos o completarlos con un valor ficticio,etc.

```python
import pandas as pd
```

```python
dt=pd.read_csv(r"C:\Users\User\Desktop\dataScience\pandasDatos\alquilerResidencial.csv",sep=";")
dt.head(10)
```

```python
#los valores nulos son los NaN, el metodo de pandas isnull() me crea un dataframe  indicando los registros que tienen un valor nulo con True.
dt.isnull()
```

```python
#el metodo notnull() hace lo opuesto, indica que valores no son nulos con un True
dt.notnull()
```

```python
#el metodo info() nos indica informacion de las columnas del df, aqui podemos saber cuantos registros de cada columna son no nulos,cuantas columnas hay en nuestro df, que columnas tienen valores nulos,etc.
dt.info()
```

```python
#la columna Valor tiene 9 valores nulos, entonces podemos buscar en el df estos registros pasando una condicion entre [dt['Valor'].isnull()] asi
dt[dt['Valor'].isnull()]
```

```python
#ahora, no es etico inventar valores para llenar estos campos nulos,entonces los vamos a borrar del df,podemos primero ver cuantos registros en total tiene nuestro df para despues verificar que si se eliminaron estos 9 valores de esta columna,para esto utilizamos el metodo shape
A=dt.shape[0]
A
```

```python
#ahora,paraeliminar estos valores nulos de esta columna Valor,utilizamos el metodo dropna() y le  indicamos la columna de la cual queremos borrar los nulos,y para que el df tome los cambios debemos pasar el parametro inplace=True,porque aqui no estamos almacenando en otra variable los resultados sino que estamos trabajando sobre el propio df,entonces para que tome los cambios se pasa el inplace.Despues puedo verificar cuantas lineas se borraron con el metodo shape.
dt.dropna(subset=['Valor'],inplace=True)
B=dt.shape[0]
A-B
```

```python
#tambien puedo verificar si esa columna tiene valores nulos con el codigo que ya teniamos
dt[dt['Valor'].isnull()]
```

Tratamiento de datos faltantes, vamos ahora a combianr columnas,por ejemplo nos piden eliminar los tipos de inmuebles que sean Departamentos y que en la columna Mantenimiento tengan valor nulo

```python
dt
```

```python
#primero miramos cuantas lineas nulas hay en la columna Mantenimiento de nuestro df
dt[dt['Mantenimiento'].isnull()].shape[0]
```

```python
#ahora,de estas 2047 filas buscamos cuales son de tipo Departamento,entonces usamos dos criterios de seleccion.Aqui se guardaran las lineas nulas,osea las que no quiero en mi df
seleccion=(dt['Mantenimiento'].isnull())&(dt['Tipo']=='Departamento')
```

```python
#Ahora,puedo contar las lineas que seran eliminadas basandome en ese criterio seleccion,simplemente hago lo que ya habiamos echo al contar las lineas del df original y restarle las lineas que quiero eliminar,el criterio es 'seleccion' y como este guarda las lineas que son nulas y yo quiero capturar las que no son nulas lo que hago es indicarle lo opuesto a este criterio con el simbolo virgulilla(~) 
A=dt.shape[0]
dt=dt[~seleccion]
B=dt.shape[0]
A-B
```

```python
#ahora, esas 745 lineas son las que quedan al restar las lineas que cumplen los dos criterios(osea las que se guardaron en la variable seleccion), pero aun quedan lineas nulas en la columna Mantenimiento,porque solo habiamos indicado las qie cumplian 2 criterios,podemos ver las que quedan nulas en Mantenimiento con el mismo codigo de antes
dt[dt['Mantenimiento'].isnull()].shape[0]
```

```python
#ahora,digamos que el cliente indica que esas filas nulas de la columna Mantenimiento y de una vez las de la columna Impuesto sean llenadas con un 0,podemos hacer esto con el metodo fillna() y como son 2 columnas podemos pasarle un diccionario que indique los valores a llenar,osea ponerles un 0.
dt=dt.fillna({'Mantenimiento':0,'Impuesto':0})
```

```python
#verificamos y debemos tener 0 nulos en Mantenimiento y en Impuesto
dt[dt['Mantenimiento'].isnull()].shape[0]
```

```python
dt[dt['Impuesto'].isnull()].shape[0]
```

```python
#con info() podemos ver como quedo nuestro df despues de borrar y rellenar los nulos en las respectivas columnas
dt.info()
```

```python
#por ultimo ya puedo exportar este nuevo df actualizado para el cliente, recordar que  aqui se reescribira el archivo que teniamos antes.
dt.to_csv('alquilerResidencial.csv',sep=';',index=False)
```
