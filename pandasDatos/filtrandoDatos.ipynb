Vamos a ver como filtrar datos con pandas para hacer busquedas mas selectivas

```python
import pandas as pd
```

```python
#traemos nuestro archivo con los datos y mostramos los primeros 10
dt=pd.read_csv(r"C:\Users\User\Desktop\dataScience\pandasDatos\PandasMateriales\data\alquiler.csv",sep=";")
dt.head(10)
```

```python
#ahora, la idea es solo mostrar los inmuebles residenciales,por lo que excluiremos los inmuebles que son comerciales,primero creamos una lista llamando solo la columna de tipo, y a esta lista le eliminamos primero los valores duplicados,la llamamos residencial
list(dt['Tipo'].drop_duplicates())
```

```python
#creamos ahora una lista llamada residencial con la lista que tenemos y a esa lista le borramos los inmuebles comerciales y la imprimimos
residencial=['Habitación',
 'Casa',
 'Departamento',
 'Casa en condominio',
 'Casa comercial',
 'Casa de villa',
 ]
residencial
```

```python
#ahora,ya teniendo esta lista residencial,podemos comparar nuestro archivo alquiler.cvs contra esta lista y usando el metodo isin() ver que elementos de alquiler.cvs concuerdan con los de la lista residencial,obtendremos una lista con booleans correspondientes.
seleccion=dt['Tipo'].isin(residencial)
seleccion
```

```python
#ahora,los datos con valor True son los que si estan en nuestra bd y concuerdan con la lista que estamos comparando de residencial,ahora ya sabemos cuales datos existen y con estos podemos crear un nuevo dataframe pasando como condicion esta lista seleccion, asi saldran solo los inmuebles residenciales,osea los que son True
datosResidencial=dt[seleccion]
datosResidencial
```

```python
#aqui solo muestra los primeros 5 y los ultimos 5 elementos de la bd,pero si queremos confirmar que efectivamente respeto la condicion de solo mostrar inmuebles residenciales podemos hacer de nuevo un drop_duplicates() pero con este nuevo dataframe,vermos que respeta el tipo de inmueble que queremos mostrar
list(datosResidencial['Tipo'].drop_duplicates())
```

```python
#para mostrar los elementos con indice consecutivo podemos utilizar como ya habiamos visto el atributo index y el metodo range,porque si vemos el dataframe resultante nos muestra los elementos pero con sus indices originales,no estan en secuencia,para mostrarlos en secuencia primero obtenemos la longitud de este df
datosResidencial.shape[0]
```

```python
#sabiendo la longitud del df ahora le aplicamos un range para que cree los indices secuenciales
datosResidencial.index=range(datosResidencial.shape[0])
datosResidencial
```

Exportando la BD

```python
#ahora, ya tenemos nuestros resultados en este nuevo df, ahora debemos exportarlo a un archivo csv para poder compartirlo,etc,entonces asi como pandas tiene un metodo para importar archivos,tambien tiene metodos para exportarlos,en este caso si queremos exportar nuestro df a un archivo csv utilizamos el metodo to_csv() se le pasa un nombre de archivo y un separador, veremos que nos crea en la misma ruta este nuevo archivo.
datosResidencial.to_csv('alquilerResidencial.csv',sep=';')
```

```python
#podemos ahora trabajar con este nuevo archivo,podemos invocarlo con un nuevo nombre de variable
datosResidencial2=pd.read_csv('alquilerResidencial.csv',sep=';')
datosResidencial2
```

```python
#vemos que se genera una columna llamada unnamed, esta columna es la de los indices del df, el metodo to_csv exporta toda la informacion,incluidos los indices,y entonces estos en el nuevo archivo generado se colocan como si fuera otra columna,para evitar que se exporten los indices se debe colocar un tercer argumento llamado index=False, asi no se creara esa columna de mas, si ejecutamos de nuevo el codigo para exportar con este nuevo parametro,no creara un nuevo archivo sino que lo reescribe.
datosResidencial.to_csv('alquilerResidencial.csv',sep=';', index=False)
```
