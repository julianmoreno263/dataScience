Vamos a ver como crear nuevas variables,por ejemplo nuestro cliente de inmuebles nos pide mostrar ademas de los valores que tenemos en nuestro df los valores brutos por metros cuadrados,el valor bruto por metro cuadrado y el valor bruto general.

Creando Nuevas Variables

```python
#trabajaremos nuevamente con nuestro archivo alquilerResidencial
import pandas as pd
```

```python
dt=pd.read_csv(r"C:\Users\User\Desktop\dataScience\pandasDatos\alquilerResidencial.csv",sep=";")
dt.head(10)
```

```python
#el valor bruto es el valor total de la suma entre Valor,Mantenimiento y de Impuesto de cada registro,para esto debemos crear una nueva variable en nuestro df.Lo hacemos asi:
dt['Valor Bruto']=(dt['Valor'])+(dt['Mantenimiento'])+(dt['Impuesto'])
dt.head(10)
```

```python
#ahora, crear el valor por metro cuadrado, para esto debemos dividir el Valor/Area,podemos redondear a dos decimales con round(2)
dt['Valor m2']=(dt['Valor'])/(dt['Area'])
dt['Valor m2']=dt['Valor m2'].round(2)
dt.head(10)
```

```python
#ahora calculamos el valor  bruto por metro cuadrado
dt['Valor Bruto m2']=((dt['Valor Bruto'])/(dt['Area'])).round(2)
dt.head(10)
```

```python
#ahora, nos piden agrupar los inmuebles que tengan la palabra Casa, y todo lo demas que se llame Departamento en una nueva columna.Entonces primero creamos una lista con estos strings,luego creamos la nueva variable que tendra como criterio la columna Tipo porque en esta columna es donde aplicaremos el calculo, para aplicar funciones o criterios en una columna utilizamos el metodo apply(),y dentro de este pasamos una funcion lambda, esta funcion tendra una variable x que representara a cada fila de la columna Tipo y le decimos que por cada fila de x le aplique la palabra 'Casa' si encuentra uno de los strings de la lista que creamos 'casa' por medio de un condicional if asi:
casa=['Casa','Casa en condominio', 'Casa de villa']
dt['Tipo Agrupado']=dt['Tipo'].apply(lambda x:'Casa' if x in casa else 'Departamento')
dt
```

Eliminando Variables

El cliente se dio cuenta que la columna impuesto esta en su mayoria con ceros, por lo cual los calculos de las otras columnas de Valor Bruto, Valor m2 y Valor Bruto m2 no quedan bien, entonces debemos excluir estos datos.

```python
#creamos un nuevo df pero solo con las columnas nuevas que agregamos.
datosAux=pd.DataFrame(dt[['Tipo Agrupado','Valor m2','Valor Bruto','Valor Bruto m2']])
datosAux.head(10)
```

```python
#para borrar una columna lo hacemos con el metodo del
del datosAux['Valor Bruto']
datosAux.head(10)
```

```python
#otra forma d eeliminar es con el metodo pop(), al ejecutarlo este metodo nos muestra lo que acabamos de eliminar
datosAux.pop('Valor Bruto m2')
```

```python
#confirmamos que ya elimino esa columna
datosAux.head(10)
```

```python
#otra forma d eeliminar es usando el metodo drop(), para verlo vamos a eliminar dos columnas de nuestro df original, debo pasarle el parametro axis=1 para indicarle que eliminare columnas,si no lo hago el por defecto piensa que debe eliminar filas y tratara de eliminarlas.
dt.drop(['Valor Bruto','Valor Bruto m2'], axis=1,inplace=True)
dt.head(10)
```

```python
#Listo, ya actualizamos nuestro df y podemos exportarlo
dt.to_csv('alquilerResidencial.csv',sep=';',index=False)
```
